# 独热表示
独热表示（One-Hot Representation）是文本表示词的一般方法，表示起来非常简单、直观，即根据单词的数量V生成一个V维的向量。在向量中只有一位为１，其他位都为０，这个向量就代表了相对应的单词。例如：将“我喜欢吃苹果，他也喜欢吃”这句话用独热表示方法表示。首先得到一个字典：{1：“我”，2：“喜欢”，3：“吃”，4：“苹果”，5：“他”，6：“也”}。其次，将我表示为[1,0,0,0,0,0]，向量共6维，即这句话包含6个不同的词，第一维度为1表示“我”在这句话中处于第一的位置。同理也可得：喜欢表示为[0,1,0,0,0,0]，吃表示为[0,0,1,0,0,0]，以此类推。

独热表示不需要学习过程，简单高效，在信息检索和自然语言处理中得到广泛的应用。如果使用独热表示方法去表示一篇文章，将这篇文章的词全部表示出来则可能需要一个成千上万维的向量，而且这个向量非常稀疏，只有特定的位置为1，其余全为0。这样的表示方法会造成维数灾难，计算起来十分困难。在语义表示上，使用独热表示来表示的词不具备语义信息，每个词的向量都是正交的，不能通过计算距离来表示两个词的相似度，所以任意两个词之间是割裂的，无法表示词与词、词与上下文之间的依赖关系，从词向量上看不出两个词是否有关联。

# 词袋模型
词袋模型（Bag of Words，BOW）[13]用文档中的每个词语的计数形成的频率向量去表示文档。假定对于一个文档，忽略了词语的前后顺序，文档中每个词的出现都是独立的，不依赖其它词是否出现。把每篇文档都看成一个装满词的袋子，然后判断这个袋子里装的都是哪些词汇，将其分类。比如文档中包含“大学”、“学术”、“论文”、“老师”等这样的词汇多，而“股市”、“汇率”、“人民币”、“走势”等这样的词汇很少，可以倾向地认为这是一篇描绘“教育”的文档，而不是描述“经济”的文档。例如：
文档1：我喜欢吃苹果，他也喜欢吃
文档2：他喜欢唱歌，我不喜欢
可以建立一个字典：{1：“我”，2：“喜欢”，3：“吃”，4：“苹果”，5：“他”，6：“也”，7：“唱歌” ，8：“不喜欢”}，字典共有9个不同的词，所以利用词袋模型表示的向量具有个9维度，每个维度代表相应的词出现的次数。可以得出：
文档1表示为：[1,2,2,1,1,1,0,0]。
文档2表示为：[1,1,0,0,1,0,1,1]。
从上可以看出，词袋模型可以将每个文档表示为一个向量，将向量和字典结合能够得到文档的语义信息，但是向量没有表达出词语在原文档中出现的顺序，所以词袋模型的表示方法损失了相邻单词的相关性和语义关系。

# CBOW
语言模型的评价标准是：语言模型能否正确预测下一个出现的词语。因此它所能利用的信息仅仅只是己经观察到的词语，
而一个用于训练词向量的模型则不受到这方面的限制，模型可以利用观察到的词语和未观察到的词语来训练得到词语的向量表示。
Mikolov利用目标词语w_t前后各n个词语来进行预测，此方法称为连续词袋模型。
